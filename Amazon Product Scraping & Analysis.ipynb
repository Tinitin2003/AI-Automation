```python
# in Reviews column remove brackets and convert to float
def parse_reviews(val):
    if pd.isna(val):
        return np.nan
    val = str(val).strip("()")  # remove parentheses
    val = val.replace(",", "").upper()  # remove commas, ensure uppercase
    if 'K' in val:
        return float(val.replace('K', '')) * 1000
    elif 'M' in val:
        return float(val.replace('M', '')) * 1_000_000
    else:
        try:
            return float(val)
        except:
            return np.nan

```

```python
df['Reviews'] = df['Reviews'].apply(parse_reviews).astype('Int64')
```

```python
df['Rating'] = df['Rating'].str.extract(r'(\d+(?:\.\d+)?)')[0].astype(float)
```

```python
df['Selling Price'] = df['Selling Price'].replace('[,,]', '', regex=True).astype(float)
```

```python
# Convert data types
df['Price'] = df['Selling Price'].astype(float)
df['Reviews'] = df['Reviews'].astype(int)
df['Rating'] = df['Rating'].astype(float)
```

# Analysis

## 1. Brand Performance Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Brand Frequency (Count of appearances)
brand_freq = df['Brand'].value_counts().reset_index()
brand_freq.columns = ['Brand', 'Frequency']

# Top 5 brands by frequency
top5_brands = brand_freq.head(5)

plt.figure(figsize=(8, 5))
sns.barplot(data=top5_brands, x='Frequency', y='Brand', palette='mako')
plt.title('Top 5 Brands by Frequency in Sponsored Products')
plt.xlabel('Number of Products')
plt.ylabel('Brand')
plt.tight_layout()
plt.show()

# 2. Average Rating by Brand (for brands with at least 5 products to avoid noise)
brand_rating = df.groupby('Brand').agg(
    Avg_Rating=('Rating', 'mean'),
    Count=('Rating', 'count')
).reset_index()

brand_rating_filtered = brand_rating[brand_rating['Count'] >= 5]

# Sort by average rating descending
top_rated_brands = brand_rating_filtered.sort_values('Avg_Rating', ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(data=top_rated_brands, x='Avg_Rating', y='Brand', palette='viridis')
plt.title('Top Rated Brands (At Least 5 Products)')
plt.xlabel('Average Rating')
plt.ylabel('Brand')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()

# 3. Pie chart for percentage share of top 5 brands by frequency
plt.figure(figsize=(7,7))
plt.pie(top5_brands['Frequency'], labels=top5_brands['Brand'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('mako', 5))
plt.title('Market Share of Top 5 Brands by Number of Sponsored Products')
plt.tight_layout()
plt.show()
```

## 2. Price vs Rating Analysis

```python
# Create rating ranges (e.g., 0-1, 1-2, ..., 4-5)
rating_bins = [0, 1, 2, 3, 4, 5]
rating_labels = ['0-1⭐', '1-2⭐', '2-3⭐', '3-4⭐', '4-5⭐']

df['Rating Range'] = pd.cut(df['Rating'], bins=rating_bins, labels=rating_labels, include_lowest=True)


import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Group and plot
avg_price = df.groupby('Rating Range')['Price'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_price, x='Rating Range', y='Price', palette='viridis')
plt.title('Average Price by Rating Range')
plt.ylabel('Average Price (₹)')
plt.xlabel('Rating Range')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Rating', y='Price', hue='Price', palette='coolwarm', size='Reviews', sizes=(20, 100))
plt.title('Price vs. Rating Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Price (₹)')
plt.legend(title='Price & Reviews', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Define high rating and low price thresholds
high_rating_threshold = 4.3
low_price_threshold = df['Price'].quantile(0.25)  # Cheapest 25%

# Good value for money
good_value = df[(df['Rating'] >= high_rating_threshold) & (df['Price'] <= low_price_threshold)]
print("✅ Good Value Products (High Rating, Low Price):")
print(good_value[['Title', 'Selling Price', 'Rating']].head(10))  # Replace with actual product column name

# Overpriced & low rated
overpriced_bad = df[(df['Rating'] <= 3.0) & (df['Price'] >= df['Price'].quantile(0.75))]
print("\n⚠️ Overpriced but Low Rated Products:")
print(overpriced_bad[['Title', 'Selling Price', 'Rating']].head(10))
```

## 3. Review & Rating Distribution

### Top 5 Products by Number of Reviews

```python
# Top 5 products by Reviews
top_reviews = df.sort_values(by='Reviews', ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(x='Reviews', y='Title', data=top_reviews, palette='Blues_r')
plt.title('Top 5 Products by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

### Top 5 Products by Rating

```python
# Filter products with at least some minimum number of reviews to avoid bias
min_reviews = 50
filtered_df = df[df['Reviews'] >= min_reviews]

# Top 5 products by Rating
top_rated = filtered_df.sort_values(by='Rating', ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(x='Rating', y='Title', data=top_rated, palette='Greens_r')
plt.title(f'Top 5 Products by Rating (with at least {min_reviews} reviews)')
plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```
